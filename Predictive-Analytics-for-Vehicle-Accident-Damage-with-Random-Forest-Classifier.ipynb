Mengubah kerusakan menjadi integer 

```python
def kerusakan_kategori(harga):
    if type(harga) != str:
        return harga
    if "OVER" in harga:
        return 2  # Kerusakan tinggi
    elif "LESS" in harga:
        return 0  # Kerusakan rendah
    else:
        return 1  # Kerusakan sedang
```

```python
# Terapkan fungsi kategori
data_train["kerusakan_kategori"] = data_train["kerusakan"].apply(kerusakan_kategori)
data_train.drop(['kerusakan'], axis=1, inplace= True)
```

```python
import pandas as pd

def onehot_encode(data, columns, prefixes):
    data = data.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(data[column], prefix=prefix)
        data = pd.concat([data, dummies], axis=1)
        data = data.drop(column, axis=1)
    return data
# fungsi one hot, 3 para meter = data frame, kolom, dan prefixes  yang akan di letakkan pada get dummy
# mengkopi data frame
# untuk setiap kolom prefix akan dipasangkan oleh zip menjadi tupple
# membuat dummies menggunakan pandas, yang akan membuatkan nilai dummy untuk setiap kolom yang diberikan (include prefix)
# menggabungkan data original dengan dummy
# menghilangkan data originallnya
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_resampled, y_resampled, train_size=0.8, random_state = 100 )
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
import optuna

def objective(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 300)
    criterion = trial.suggest_categorical("criterion", ['gini', 'entropy', 'log_loss'])
    max_depth = trial.suggest_int("max_depth", 1, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 30)

    clf = RandomForestClassifier(n_estimators=n_estimators,
                                criterion = criterion, #
                             max_depth = max_depth,
                             min_samples_split = min_samples_split,
                             random_state = 42)
    
    clf.fit(X_train, y_train)
    y_prediksi = clf.predict(X_test)

    score = f1_score(y_test, y_prediksi, average="micro")
    return score


study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)

best_params = study.best_params
```

```python

clf = RandomForestClassifier(
        **best_params
    )

clf.fit(X_train, y_train)
```
